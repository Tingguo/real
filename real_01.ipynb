```python
# !pip install BeautifulSoup4
import requests
import bs4,re

import urllib
import json
import codecs

import pandas as pd
import numpy as np

import datetime
```

```python
# get data from 2012-01-01 to 2013-01-01
start = datetime.datetime(2012,1,1)
end = start.replace(start.year + 1)
print(start.strftime("%Y-%m-%d"))
print(end.strftime("%Y-%m-%d"))

loc = '"localities":[{"locality":"waterloo","subdivision":"nsw","postcode":"2017"}]'
basic = '"channel":"sold","pageSize":"200"'
f = '"filters":{"soldDateRange":{"minimum":"%s","maximum":"%s"}}' % (start.strftime("%Y-%m-%d"),end.strftime("%Y-%m-%d"))
url = 'https://services.realestate.com.au/services/listings/search?query={%s,%s,%s}' %(loc,basic,f)

json_string = requests.get(url).text
data = json.loads(json_string)
total_number = data['totalResultsCount']
total_pages = np.ceil(total_number/200.0)
print('total pages is: %d' % total_pages)

print(url)
```

```python
def get_json(url):
    json_string = requests.get(url).text
    data = json.loads(json_string)
    return data
```

```python
results = pd.DataFrame()
propertyId = []
address = []
features = []
description = []
propertyType = []
dateSold = []
agency = []
price = []
listers = []
title = []
signature = []

for i in range(1,int(total_pages+1)):
    page = '"page":"%d"' % i
    url = 'https://services.realestate.com.au/services/listings/search?query={%s,%s,%s,%s}' %(loc,basic,page,f)
    data = get_json(url)
    print(i)
    for j in range(len(data['tieredResults'])):
        print('%d-th number: %d' % (j, len(data['tieredResults'][j]['results'])))
        if len(data['tieredResults'][j]['results'])>0:
            for k in range(len(data['tieredResults'][j]['results'])):
                tmp = data['tieredResults'][j]['results'][k]
                if 'propertyId' in tmp.keys():
                    propertyId.append(tmp['propertyId'])
                else:
                    propertyId.append('')
                if 'address' in tmp.keys():
                    address.append(tmp['address'])
                else:
                    address.append('')
                if 'features' in tmp.keys():
                    features.append(tmp['features'])
                else:
                    features.append('')
                if 'description' in tmp.keys():
                    description.append(tmp['description'])
                else:
                    description.append('')
                if 'propertyType' in tmp.keys():
                    propertyType.append(tmp['propertyType'])
                else:
                    propertyType.append('')
                if 'dateSold' in tmp.keys():
                    dateSold.append(tmp['dateSold'])
                else:
                    dateSold.append('')
                if 'agency' in tmp.keys():
                    agency.append(tmp['agency'])
                else:
                    agency.append('')
                if 'price' in tmp.keys():
                    price.append(tmp['price'])
                else:
                    price.append('')
                if 'listers' in tmp.keys():
                    listers.append(tmp['listers'])
                else:
                    listers.append('')
                if 'title' in tmp.keys():
                    title.append(tmp['title'])
                else:
                    title.append('')
                if 'signature' in tmp.keys():
                    signature.append(tmp['signature'])
                else:
                    signature.append('')
results['propertyId'] = propertyId
results['address'] = address
results['features'] = features
results['description'] = description
results['propertyType'] = propertyType
results['dateSold'] = dateSold
results['agency'] = agency
results['price'] = price
results['listers'] = listers
results['title'] = title
results['signature'] = signature

results.to_csv('waterloo_2012_2013.csv',index=False)

```
